```python
# Librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

```python
# Subir base de Datos
df = pd.read_csv('inmuebles2.csv')
df.head(5)
```

```python
# Eliminar la primera fila
data1 = pd.read_csv('inmuebles2.csv', skiprows=1) # Se salta una fila, desde el inicio
data1
```

```python
# Eliminar columnas
data2= data1.drop(['0','Referencia'], axis=1) # Se eliminaran la Columnas llamadas "0" y "Referencia"
data2
```

```python
# Identificar valores nulos por columna
nulos = data2.isnull().sum() # Usa la funcion para conocer cuantos datos vacios o nulos hay por columna y hace un sumado
nulos
```

**Primer Método de Sustitución de Valores Nulos**

```python
# Copiar al DataFrame
data3 = data2.copy() # Copia el DF llamado data2
data3
```

```python
# Sustituir valores nulos con promedio o media
data3["Superficie"] = data3["Superficie"].fillna(round(data2["Superficie"].mean(),1))
```

```python
# Mostrar nuevamente los valores nulos
nulos = data3.isnull().sum()
nulos

# Notar que en "Superficie" ya no hay valores nulos
```

**Segundo Método de Sustitución de Valores Nulos**

```python
#Sustituir valores nulos con mediana
data3["Precio Venta"]=data3["Precio Venta"].fillna(round(data2["Precio Venta"].median(),1))
data3
```

```python
# Corroborar los cambios
nulos=data3.isnull().sum()
nulos

# Notar que "Superficie" y "Precio Venta" sale 0
```

**Tercer Método de Sustitución de Valores Nulos**

```python
#Sustituir valores nulos por un valor numérico en concreto
data3["Fecha Alta"]= data3["Fecha Alta"].fillna(10)
data3
```

```python
#Corroborar Valores Nulos
nulos=data3.isnull().sum()
nulos

# Notar que "Fecha Alta" sale con 0
```

**Cuarto Método de Sustitución de Valores Nulos**

```python
#Sustituir valores nulos por un string en  concreto
data3["Fecha Venta"] =data3["Fecha Venta"].fillna("2030-06-19 00:00:00")
```

```python
#Corroboramos valores nulos
nulos=data3.isnull().sum()
nulos

# Notar que "Fecha Venta" aparece ahora con 0
```

**Quinto Método de Sustitución de Valores Nulos**

```python
#Sustituir valores nulos por valores no nulos hacia adelante "forward fill" ("ffill")
data3["Tipo"] =data3["Tipo"].fillna(method="ffill")
```

```python
#Corroborar Valores Nulos
nulos=data3.isnull().sum()
nulos

# Notar que "Tipo" sale con 0
```

**Sexto Método de Sustitución de Valores Nulos**

```python
#Sustituir valores nulos por valores no nulos hacia atrás backward fill" ("bfill")
#Filtro por columnas
data3["Operación"] =data3["Operación"].fillna(method="bfill")
```

```python
#Corroborar Valores Nulos
nulos=data3.isnull().sum()
nulos

# Notar que ahora "Operación" sale con 0
```

```python
# Notar que, tanto "Provincia" y "Vendedor" siguen con datos nulos, por lo que se hará un relleno a ambos

# Sustituir valores nulos por valores no nulos hacia atrás backward fill" ("bfill")
data3["Provincia"] =data3["Provincia"].fillna(method="bfill")

# Sustituir valores nulos por valores no nulos hacia adelante forward fill" ("ffill")
data3["Vendedor"] =data3["Vendedor"].fillna(method="ffill")
```

```python
# Corroborar valores nulos
nulos=data3.isnull().sum()
nulos

# Notar que, por fin, ya no hay datos nulos
```

```python
# Corroboramr valores nulos en todo el dataframe
# Ya no hay filtro por columnas
nulos=data3.isnull().sum().sum()
nulos
```

```python
# Las librerias ya estan importadas hasta arriba

# Base de Datos
VT = pd.read_csv("Ventas_totales.csv")
VT
```

```python
# Identificar valores nulos por columna
nulos = VT.isnull().sum()
nulos
```

```python
# Copiar el DataFramce
data = VT.copy()
data
```

```python
# Sustituir valores nulos con promedio o media
data["salon_ventas"] = data["salon_ventas"].fillna(round(VT["salon_ventas"].mean(),1))
data["tarjetas_debito"] = data["tarjetas_debito"].fillna(round(VT["tarjetas_debito"].mean(),1))
data["tarjetas_credito"] = data["tarjetas_credito"].fillna(round(VT["tarjetas_credito"].mean(),1))

# Mostrar los valores nulos
nulos = data.isnull().sum()
nulos

# Notar que en "Salon_ventas", "Tarjetas_debito" & "Tarjetas_credito" ya no hay valores nulos
```

```python
# Sustituir valores nulos con mediana
data["otros_medios"] = data["otros_medios"].fillna(round(VT["otros_medios"].median(),1))
data["subtotal_ventas_alimentos_bebidas"] = data["subtotal_ventas_alimentos_bebidas"].fillna(round(VT["subtotal_ventas_alimentos_bebidas"].median(),1))
data["bebidas"] = data["bebidas"].fillna(round(VT["bebidas"].median(),1))

# Mostrar los valores nulos
nulos = data.isnull().sum()
nulos

# Notar que "Otros_medios", "Subtotal_ventas_alimentos_bebidas" & "Bebidas" salen con 0
```

```python
# Sustituir valores nulos por un valor numérico en concreto
data["almacen"] = data["almacen"].fillna(10)
data["panaderia"] = data["panaderia"].fillna(10)
data["lacteos"] = data["lacteos"].fillna(10)

# Mostrar valores nulos
nulos = data.isnull().sum()
nulos

# Notar que "Almacen", "Panaderia" & "Lacteos" salen con 0
```

```python
# En esta parte iría la parte de reemplazar valores nulos con String, sin embargo se usa una Base de Datos de valores enteros (int),
# por ello se saltará este código de sustitución

# Sustituir valores nulos por valores no nulos hacia adelante "forward fill" ("ffill")
data["carnes"] = data["carnes"].fillna(method = "ffill")
data["verduleria_fruteria"] = data["verduleria_fruteria"].fillna(method = "ffill")
data["alimentos_preparados_rotiseria"] = data["alimentos_preparados_rotiseria"].fillna(method = "ffill")

# Mostrar los valores nulos
nulos = data.isnull().sum()
nulos

# Notar que "Carnes", "Verduleroa_fruteria" & "Alimentos_preparados_rotiseria" salen con 0
```

```python
# Sustituir valores nulos por valores no nulos hacia atrás backward fill" ("bfill")
data["indumentaria_calzado_textiles_hogar"] = data["indumentaria_calzado_textiles_hogar"].fillna(method = "bfill")
data["electronicos_articulos_hogar"] = data["electronicos_articulos_hogar"].fillna(method = "bfill")
data["otros"] = data["otros"].fillna(method ="bfill")

# Corroborar Valores Nulos
nulos = data.isnull().sum()
nulos

# Notar que "Indumentaria_calzado_textiles_hogar", "Electronicos_articulos_hogar" & "Otros" salen con 0
```
